```python
"""
    Deividi Luiz Pinto da Silva
    Universidade Federal da Fronteira Sul - UFFS Champus Chapecó
    Semestre 2021/2
    
    Segurança e auditoria de sistemas

    ####### INSTRUÇÕES E OBSERVAÇÕES ##########
    
    Como informar chave privada: p q d
                    EX: 11 22 33
                    Não importam os valores, porém precisa deixar um espaço separando para realizar a separação dos valores
    
    Como informar chave pública: p q
                    EX: 11 22
                    Não importam os valores, porém precisa deixar um espaço separando para realizar a separação dos valores
                    
    Para ler um arquivo não é necessário informar a extensão '.txt', o programa já entende e faz a leitura correta.
    
    Caso tentar ler um arquivo com a chave privada incorreta, poderá ser descifrado, porém incorretamente, visto que para ter acesso
        a mensagem original, precisará das chaves corretas.
        
    Caso o passo acima ocorrer, um novo arquivo é gerado (com nome 'descrifrado' ao final, porém o antigo permanece cifrado, podendo
        assim tentar descrifrar novamente.
        
    ###########################################
"""

import random #importação da biblioteca para gerar um valor aleatório
import string #importação da biblioteca para trabalhar com strings e caracteres

# import totient() method from sympy
from sympy.ntheory.factor_ import totient

#alfabeto que será usado, considerando mais caracteres
#pode-se adicionar quantas e qualquer caracteres necessários para criptografar na mensagem

alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z', ' ', '.', '?', '!']

#função para descrifrar a mensagem
def describe():
    name_archive = input('Informe o nome do arquivo para descifrar: ') #recebe o nome do arquivo informado pelo usuário
    name_archive_font = name_archive #salva o nome fonte
    name_archive+='.txt' #adiciona extensão do arquivo ao final .txt
    
    try: #tenta executar, caso ocorrer algum erro durante a execução, vai para o except
        with open(name_archive, "r") as tf: #abre o arquivo
            lines = tf.read().split('\\') #le as linhas
            
        list_encrypt = [] #lista que receberá cada caractere cifrado
    
        private_key = str(input('Informe a chave privada. Formato > P Q D <:')) #recebe a chave privada informada pelo usuário em modo string
        private_key = private_key.split(' ') #separa em cada posição retirando o espaço
        
        while len(private_key) != 3: #enquanto o tamanho for diferente de 3 será solicitado nova operação ao usuário
            private_key = str(input('Informe a chave privada novamente: ')) #chave privada novamente
            private_key = private_key.split(' ') #separa novamente

        p = int(private_key[0]) #p recebe a primeira posição da lista com a chave privada
        q = int(private_key[1]) #q recebe a segunda posição da lista com a chave privada
        d = int(private_key[2]) #d recebe a terceira posição da lista com a chave privada

        n = p*q #n realiza o calculo multiplicando os numeros primos p e q

        for count in range(0, len(lines)): #percorre toda a linha que recebeu o texto
            if lines[count] != '': #se a linha nao for vazia
                list_encrypt.append(lines[count]) #lista cifrada recebe o caractere

        list_describe = [] #lista que irá receber o caractere descifrado

        for count in range(0, len(list_encrypt)): #percorre toda a lista cifrada
            character = int(list_encrypt[count]) ** d % n #cada caractere receberá sua cifra original após realizar o cálculo
            list_describe.append(character) #adiciona na lista descifrada
        
        describe_message = "" #limpa o buffer
        chacter_describe = [] #lista do caractere descifrado

        for count in range(0, len(list_describe)): #percorre toda a lista descifrada
            for count2 in range(0, len(alphabet)): #percorre todo o alfabeto definido no topo
                if list_describe[count] == count2: #se encontrar o caractere igual ao do alfabeto
                    chacter_describe.append(alphabet[count2]) #adiciona o caractere do alfabeto na lista de caractere descifrado
                    describe_message += alphabet[count2] #mensagem descrifrada recebe a letra do alfabeto
        
        name_archive_font+=' descifrado.txt' #nome do arquivo fonte concatena a placa 'descifrado' ao final, gerando um arquivo novo
        archive = open(name_archive_font, 'w') #cria e abre o arquivo
        archive.writelines(str(describe_message)) #escreve a mensagem completa
        archive.close() #fecha o arquivo
        print('\nNovo arquivo criado, para ler a mensagem acesse o arquivo: ', name_archive_font) #informa ao usuário o nome do novo arquivo gerado
    except: #caso ocorrer algum problema na execução
        print('\nErro: Não foi possível descifrar, tente novamente.') #informa ao usuário que ocorreu um problema
        return

#função para cifrar a mensagem
def encrypt():
    list_encrypt = [] #lista para cifrar
    list_original = [] #lista original de cada caractere
    save_message = "" #limpa o buffer
    
    public_key = str(input('Informe a chave pública. Formato > P Q <:')) #solicita ao usuário a chave pública
    public_key = public_key.split(' ') #separa em cada posição retirando o espaço
    
    while len(public_key) != 2: #enquanto o tamanho for diferente de 2 será solicitado nova operação ao usuário
        public_key = str(input('Informe a chave privada novamente: '))
        public_key = public_key.split(' ') #separa novamente

    n = int(public_key[0]) #n recebe a primeira posição da lista com a chave pública
    e = int(public_key[1]) #e recebe a segunda posição da lista com a chave pública
    
    for count in range(0, len(message)): #percorre toda a mensagem
        for count2 in range(0, len(alphabet)): #perocrre todo o alfabeto
            if message[count] == alphabet[count2]: #se o caractere da mensagem for igual ao que pertence ao alfabeto

                list_original.append(count2) #lista original recebe cada caractere
                save = count2 ** e % n #save recebe o valor cifrado com base na chave pública informado pelo usuário
                list_encrypt.append(save) #adiciona na lista de cifras
        
    name_archive = input('Informe o nome do arquivo para cifrar: ') #solicita um nome ao arquivo para o usuário
    name_archive+='.txt' #adiciona extensão .txt
    
    try: #tenta executar, caso ocorrer algum erro durante a execução, vai para o except
        archive = open(name_archive, 'w') #cria e abre o novo arquivo
        print("Arquivo cifrado sucesso.") #informa ao usuário
        
        save_message+="\\" #adiciona caractere \ para poder saber quais caracteres precisarão ser cifrados e descifrados
        for count in range(0, len(list_encrypt)): #percorre toda a lista cifrada
            save_message += str(list_encrypt[count])+'\\' #adiciona após cada caractere a \ (contra-barra) indicando o fim do caractere
        
        archive.writelines(str(save_message)) #escreve a mensagem completa e cifrada com \ no arquivo
        archive.close() #fecha o arquivo
        
    except: #caso ocorrer algum problema na execução
        print('\nErro: Arquivo não localizado.')

#função define primos        
def define_primes(totient_n): #recebe o totiente do número n
    first_list = [] #primeira lista
    second_list = [] #segunda lista
    list_primes = [] #terceira lista

    init = 2 #define o inicial em 2
    primes_like = False #define o primos entre si como Falso

    for count in range (1, totient_n+1): #percorre toda a quantidade de totient_n
        if (totient_n % count == 0): #se o mod do totient com o contador for resto zero, adiciona na primeira lista
            first_list.append(count)

    while (init > 1 and init < totient_n): #enquanto o inicial for maior que 1 e menor que o totient
        for count in range (1, init+1): #percorre realizando a verificação se possui resto zero
            if (init % count == 0): #se inicial resto contador for zero
                second_list.append(count) #adiciona na segunda lista

        for count in range (0, len(first_list)): #percorre toda a primeira lista
            for count2 in range (0, len(second_list)): #percorre toda a segunda lista
                if first_list[count] == second_list[count2]: #se os números tiverem o mesmo divisor
                    if first_list[count] != 1: #verifica se o único divisor em comum é 1
                        primes_like = False #se não for retorna False
                        break
                    else:
                        primes_like = True #se for retorna Verdadeiro indicando que são primos entre si
        if primes_like == True: #se forem primos entre si, ou seja, único divisor comum dos valores é 1
            list_primes.append(init) #adiciona o valor de inicial (atual) na lista de primos entre si

        primes_like = False #define novamente para Falso
        second_list = [] #limpa a lista
        init+=1 #inicial soma 1

    e = random.randint(0,len(list_primes)) #e vai definir um valor aleatório da lista de primos entre si, que vai de 0 até o tamanho total da lista
    e = list_primes[e] #e vai receber o valor da lista

    d = pow(e, -1, totient_n) #d vai receber o valor do inverso multiplicativo de 'e' em (mod phi(n)) => phi(n) é o totient_n

    return e, d #retorna 'e' e 'd'

#função para saber se o número informado é primo
def is_prime(n):
    mult = 0 #variavel de contagem de divisão
    
    for count in range(2,n): #percorre de 2 até o valor solicitado
        if (n % count == 0): #calcula o resto da divisão
            mult += 1 #se for zero soma 1 em mult

        if(mult==0): #caso o mult não encontre nenhum resto zero
            return True #retorna Verdadeiro
        else:
            print('Erro: O número {} não é primo, escolha outro.'.format(n)) #caso Falso, é informado ao usuário que o número não é primo
            return False

#função MAIN (principal) responsável por chamar as demais funções quando necessário
if __name__=='__main__':

    option = 0 #opção recebe o valor zero de inicio
    message = "" #mensagem é limpa o buffer
    
    prime = False #primo recebe Falso
    
    while (prime == False): #enquanto o valor informado não for primo
        p = input('Informe o 1º número primo: ') #solicita o primeiro número primo
        
        if p.isnumeric() != True: #enquanto o valor informado for inválido será solicitado novamente
            print('Erro: O valor informado não é válido.')
        else: #se for válido
            p = int(p) #converete para int o valor de p
            prime = is_prime(p) #chama a função para testar se o número é primo
 
    prime = False #primo volta a ser Falso
    
    while (prime == False): #enquanto o valor informado não for primo
        q = input('Informe o 2º número primo: ') #solicita o segundo número primo
        
        if q.isnumeric() != True: #enquanto o valor informado for inválido será solicitado novamente
            print('Erro: O valor informado não é válido.')
        else: #se for válido
            q = int(q) #converete para int o valor de q
            prime = is_prime(q) #chama a função para testar se o número é primo

    n = p * q #realiza o cálculo de n
    totient_n = totient(n)  #totient recebe o valor chamando a função para calcular o totient de n => totient = totiente de Euler
    
    e, d = define_primes(totient_n) #'e' e 'd' recebem os valores de retorno da função para definir os primos  
    
    print('\nChave pública: ', n, e) #informa ao usuário a chave pública gerada
    print('Chave privada: ', p, q, d) #informa ao usuário a chave privada gerada
    
    #vai repetir até que seja interrompido
    while (True):        
        print('\n#MENU#\n\n1 - Digitar mensagem\n'
               '2 - Cifrar mensagem\n'
               '3 - Descifrar mensagem\n'
               '4 - Mostrar números primos escolhidos\n'
               '5 - Mostrar chaves\n'
               '6 - Ler mensagem do arquivo\n'
               '0 - Sair\n\n')
        
        option = input('Opcao: ')
        if option.isnumeric() != True:
            print('Erro: Opção inválida! Por favor selecione novamente.')
        else:
            option = int(option)
        
        if option == 1:  #opção numero 1 => para digitar a mensagem e criar o arquivo
            message = input('Digita sua mensagem: ')
            
            name_archive = input('Digite o nome do arquivo para salvar: ')
            name_archive+='.txt'
            
            archive = open(name_archive, 'w')
            
            archive.writelines(str(message))
            archive.close()
            
        elif option == 2: #opção numero 2 => para cifrar a mensagem do arquivo
            if message != "": 
                encrypt()
                message = ""
            else:
                print('Por favor, primeiro informe uma mensagem para cifrar.')
        elif option == 3: #opção numero 3 => para descrifrar a mensagem do arquivo
            describe()
        elif option == 4:  #opção numero 4 => para msotrar os números primos escolhidos
            print('\nNumeros primos escolhidos: {}, {} '.format(p, q))
        elif option == 5: #opção numero 5 => para mostrar as chaves geradas
            print('\nChave pública: ', totient_n, e)
            print('Chave privada: ', p, q, d)
        elif option == 6: #opção numero 6 => para ler o arquivo desejado
            name_archive = input('Informe o nome do arquivo para ler: ')
            name_archive+='.txt'
            try:
                with open(name_archive, "r") as tf:
                    lines = tf.read()
                    print('\nMensagem: ')
                    print(lines)
            except:
                print('\nErro: Arquivo não localizado.')
        elif option == 0: #opção numero 0 => para interromper a execução do programa
            break
```
